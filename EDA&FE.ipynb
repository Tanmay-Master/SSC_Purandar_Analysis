```python
import pandas as pd
import os
import re

# List of your files (adjust the folder path if needed)
files = [
    "Data/SSC2021.xlsx",
    "Data/SSC2022.xlsx",
    "Data/SSC2023.xlsx",
    "Data/SSC2024.xlsx",
    "Data/SSC2025.xlsx"
]

# Empty list to hold DataFrames
dfs = []

for file in files:
    # Extract year from filename using regex (looks for 4 digits)
    match = re.search(r'(\d{4})', file)
    if match:
        year = int(match.group(1))
    else:
        raise ValueError(f"Year not found in filename: {file}")
    
    # Read the Excel file
    df = pd.read_excel(file)
    
    # Add the year column
    df['Year'] = year
    
    # Append to list
    dfs.append(df)

# Concatenate all DataFrames
combined_df = pd.concat(dfs, ignore_index=True)

# Save to CSV (without index, and ensure commas in college names are quoted)
combined_df.to_csv('combined_ssc_data.csv', index=False)

print("CSV file created: combined_ssc_data.csv")
print(f"Total rows: {len(combined_df)}")
```
